# Basic Recommndation system using weighted average

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading Data-frame

credits= pd.read_csv('tmdb_5000_credits.csv')
movies_df= pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits.head()
```

```python
movies_df.head()
```

```python
# Checking the dimension of dataframe

print("Credits:",credits.shape)
print('movies Dataframe:',movies_df.shape)
```

```python
# Merging the movie dataframe and credits data frame based on "id"

credits_col_rename=credits.rename(index=str,columns={'movie_id':'id'})
movies_df_merged=movies_df.merge(credits_col_rename,on='id')
movies_df_merged.head()
```

```python
# Removing unnecessary columns.

movies_cleaned_df=movies_df_merged.drop(columns=['homepage','title_x','title_y','status', 'production_countries'])
movies_cleaned_df.head()
```

```python
# Checking for null values

movies_cleaned_df.info()
```

# Recommending movies by Weighted rating

```python
v=movies_cleaned_df['vote_count']
R=movies_cleaned_df['vote_average']
C=movies_cleaned_df['vote_average'].mean()
m=movies_cleaned_df['vote_count'].quantile(.70)   # Selecting movies having vote count quantile above .70
```

```python
movies_cleaned_df['weighted_average']=((R*v)+(C*m))/(v+m)
```

```python
movies_cleaned_df.head()
```

```python
# Sorting Movies based on weighted rating

movie_sorted_ranking=movies_cleaned_df.sort_values('weighted_average',ascending=False)
movie_sorted_ranking[['original_title','vote_count','vote_average','weighted_average','popularity']].head(20)
```

```python
weight_avg=movies_cleaned_df.sort_values('weighted_average',ascending=False)
sns.set_style("darkgrid")
plt.figure(figsize=(12,7))
axis1=sns.barplot(x=weight_avg['weighted_average'].head(15),y=weight_avg['original_title'].head(15),linewidth=1.5,edgecolor="black")
plt.title('Best top 15 movies by weighted average votes',weight='bold',color='b',fontsize=15)
plt.xlabel('Weighted Average',weight='bold',fontsize=15)
plt.ylabel('Movie name',weight='bold',fontsize=15)
plt.yticks(fontsize=13)
plt.grid(True)
```

# Recommending movies by Popularity

```python
popularity=movie_sorted_ranking.sort_values('popularity',ascending=False)
sns.set_style("darkgrid")
plt.figure(figsize=(12,7))
axis1=sns.barplot(x=popularity['popularity'].head(15),y=popularity['original_title'].head(15),linewidth=1.5,edgecolor="black")
plt.title('Best top 15 movies by popularity',weight='bold',color='b',fontsize=15)
plt.xlabel('popularity',weight='bold',fontsize=15)
plt.ylabel('Movie name',weight='bold',fontsize=15)
plt.yticks(fontsize=11)
plt.grid(True)
```

# Recommendation based on (Hybrid) scaled weighted avg and popularity (60:40)

why 60:40 ratio ?

* Because latest film are more popular than old films. Therefore giving 50% weitage to both is not fair.

```python
# Scaling weighted rating and popularity as they are not on same scale

from sklearn.preprocessing import MinMaxScaler

scaling=MinMaxScaler()
movie_scaled_df=scaling.fit_transform(movies_cleaned_df[['weighted_average','popularity']])
movie_normalized_df=pd.DataFrame(movie_scaled_df,columns=['weighted_average','popularity'])
movie_normalized_df.head()
```

```python
movies_cleaned_df[['norm_weighted_average','norm_popularity']]=movie_normalized_df
```

```python
movies_cleaned_df.head()
```

```python
# Calculating hybrid rating using weighted rating and popularity

movies_cleaned_df['score']=movies_cleaned_df['norm_weighted_average']*0.6 + movies_cleaned_df['norm_popularity']*0.4
movies_cleaned_df= movies_cleaned_df.sort_values(['score'],ascending=False)
movies_cleaned_df[['original_title','norm_weighted_average','norm_weighted_average','score']].head(10)
```

```python
score_df=movies_cleaned_df.sort_values('score',ascending=False)
sns.set_style("darkgrid")
plt.figure(figsize=(12,7))
axis1=sns.barplot(x=score_df['score'].head(15),y=score_df['original_title'].head(15),linewidth=1.5,edgecolor="black")
plt.title('Best rated top 15 movies by popularity and avg weightage',weight='bold',color='b',fontsize=15)
plt.xlabel('Hybrid rating',weight='bold',fontsize=15)
plt.ylabel('Movie name',weight='bold',fontsize=15)
plt.yticks(fontsize=13)
plt.grid(True)
```
